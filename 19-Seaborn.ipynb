```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.scatter(iris['sepal_length'], iris['sepal_width'])
```

```python
sns.pairplot(iris, hue='species')
```

```python
titanic = sns.load_dataset('titanic')
titanic.drop(['alone', 'alive', 'who', 'adult_male', 'embark_town', 'class', 'deck'], axis=1, inplace=True)
titanic.head()
```

```python
sns.pairplot(titanic)
```

```python
sns.catplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.boxplot(x='pclass', y='age', data=titanic, hue='sex')
```

```python
sns.distplot(titanic['fare'])
```

```python
sns.jointplot(x='age', y='fare', data=titanic, kind='kde')
```

```python
sns.heatmap(titanic.drop(['sex', 'embarked'], axis=1).corr())
```
